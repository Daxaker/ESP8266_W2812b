```python
byteArr = bytearray()
import paho.mqtt.publish as publish
from PIL import Image 
mImg = Image.open("/Users/daxaker/Downloads/mario.ico")
mImg = mImg.resize((16,16), Image.ANTIALIAS)
```

```python
for x in range(mImg.size[0]):
    for y in range(mImg.size[1]):
        rgb = bytes(mImg.getpixel((x,y)))
        byteArr.append(rgb[0])
        byteArr.append(rgb[1])
        byteArr.append(rgb[2])
```

```python
publish.single("ledMatrix/blob", bytes(byteArr))
```
